```python
import matplotlib.pyplot as plt
import numpy as np

Producto = np.array(["Producto1", "Producto2", "Producto3", "Producto4", "Producto5"])
Cantidad = np.array([15, 30, 25, 40, 20])
Precio = np.array([7.0, 8.5, 6.5, 10.0, 5.5])

precio_total = Cantidad * Precio
print(precio_total)

plt.bar(Producto, precio_total)
plt.title("Precio total por producto")
plt.savefig("grafico_precios2.png")
```
